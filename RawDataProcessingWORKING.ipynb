```python
import pandas as pd
import os
```

```python
file = 'trades2008.csv'
```

```python
# Читаем файл
os.chdir('C:/Users/sstarikov/Desktop/Новая папка/Trades')
trades = pd.read_csv(file, sep=';', skiprows = 2)

# Фильтр
uts = trades[(trades.iloc[:, 2]=='USD000UTSTOM') & (trades.iloc[:, 3]=='CETS') & (trades.iloc[:, 1]<='11:30:00')]

# Отсекаем лишние столбцы
uts1 = uts.copy()
uts1.drop(['SECID', 'BOARDID', 'INVCURVOL', 'BUYSELL', 'TRADENO'], axis = 1, inplace = True)

# Обработка дат и времени
uts1['DATETIME'] = uts1['TRADEDATE'] + ' ' + uts1['TRADETIME']
uts1['DATETIME'] = pd.to_datetime(uts1['DATETIME'], format='%d.%m.%Y %H:%M:%S')
uts1.drop(['TRADETIME'], axis = 1)
uts1 = uts1[['TRADEDATE', 'DATETIME', 'PRICE', 'VOLCUR']]

# Замена запятых на точки
comma_replace = lambda x: float(x.replace(',', '.'))
uts1['PRICE'] = uts1['PRICE'].apply(comma_replace)

# Меняем названия столбцов
uts1.rename(columns={'PRICE':'RATE', 'VOLCUR':'VOLUME'}, inplace = True)

# Сохраняем в csv
uts1.to_csv(os.path.join('C:/Users/sstarikov/Desktop/Новая папка/Trades processed', file), sep=';', index=False)

# Check
uts1.head()
```
