```python
import pandas as pd
import numpy as np
import datetime as dt
import yfinance as yf
```

```python
# Share
equity = 'AFKS.ME'

# Horizon
horizon = 250

# Historical horizon (in years)
history = 5

# Percentile
p = 0.05
```

```python
# Get historical market data
share = yf.Ticker(equity)
hist = share.history(period = 'max')['Close']
df = pd.DataFrame(hist)
df['Change'] = df['Close'] / df['Close'].shift(horizon) - 1
perc = df['Change'].quantile(p)
perc
```
