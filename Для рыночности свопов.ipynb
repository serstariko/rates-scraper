```python
import os
import pandas as pd
import glob
import webbrowser
import zipfile
```

```python
# Upload files from MOEX website for chosen dates
# Enter dates
# ddmm
Dates = ['1410', '1410', '1410', '1410', '1410']

print('Downloaded:')
for date in Dates:
    month = date[2:]
    day = date[:2]
    print(day, month)
    webbrowser.open_new('https://iss.moex.com/iss/downloads/engines/currency/markets/selt/years/2021/months/'+ 
                        month +'/days/' + day + '/trades_moex_currency_selt_2021_' + month + '_' + day + '.csv.zip')
```

```python
# Upload file
os.chdir('C:/Users/sstarikov/Documents/Рыночность свопов')

# Working with zip
List_of_zip = [i for i in glob.glob('*.{}'.format('zip'))]

for zips in List_of_zip:
    unfold_zip = zipfile.ZipFile(zips)
    unfold_zip.extractall()
    os.rename('trades.csv', 'trades' + zips[34:36] + zips[31:33] + '.csv')
    unfold_zip.close()

# Working with csv
trades_appended = pd.DataFrame()
List_of_files = [i for i in glob.glob('*.{}'.format('csv'))]

for files in List_of_files:
    trades = pd.read_csv(files, sep=';', header = 1)
    trades_filter = trades[trades.SECID.isin(['USD000TODTOM', 'CNYRUBTODTOM'])]
    trades_filter.drop(trades_filter[['BOARDID', 'BUYSELL', 'TRADENO']], axis = 1, inplace = True)
    trades_appended = trades_appended.append(trades_filter)    
    
trades_appended.to_csv('tradesALL.csv', sep = ';', index = False)
```
